```python
import json
from collections import defaultdict
```

```python
with open('../data/out/CS魔法社_10-09T22-36.json', 'r') as f:
    records = json.load(f)
```

```python
userStat = defaultdict(int)

USER_ID_SPLITER = ":\n"

for record in records[:]:
    msgContent = record['msgContent']
    if record['msgStatus'] == 4 and USER_ID_SPLITER in msgContent and not msgContent.startswith('<'):
        userId = msgContent.split(USER_ID_SPLITER)[0]
        userStat[userId] += 1
userStat = dict(sorted(userStat.items(), key=lambda x: -x[1]))
userStat
```
